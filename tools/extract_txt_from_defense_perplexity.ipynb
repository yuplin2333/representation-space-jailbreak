```python
import pandas as pd
import ast
```

```python
CSV_FILE_PERPLEXITY = f"results/defense_perplexity_ours_autodan_llama2-7b.csv"
CSV_FILE_ORIGINAL = f"results/final/ours_autodan_llama2-7b_100.csv"
OUTPUT_PREFIX = "data/visualization_defense_datasets/perplexity_ours_autodan_llama2-7b_"
```

```python
def literal_eval(s):
    try:
        return ast.literal_eval(s)
    except ValueError:
        return s
```

```python
df_perplexity = pd.read_csv(CSV_FILE_PERPLEXITY)
df_original = pd.read_csv(CSV_FILE_ORIGINAL)

# select "prompt" that "jailbroken" == True in df_original but "jailbroken" == False in df_perplexity
df_perplexity["prompt"] = df_perplexity["prompt"].apply(literal_eval)
try:
    df_original["prompt"] = df_original["prompt"].apply(literal_eval)
except:
    pass
df_perplexity["adv_prompt"] = df_perplexity["adv_prompt"].apply(literal_eval)

try:
    df_original["adv_prompt"] = df_original["adv_prompt"].apply(literal_eval)
except KeyError:
    print("No adv_prompt in original dataset")
    df_original["adv_string"] = df_original["adv_string"].apply(literal_eval)
    df_original["adv_prompt"] = df_original["prompt"] + " " + df_original["adv_string"]
except SyntaxError:
    print("original dataset adv_prompt is not repr")
    pass

merged_df = pd.merge(
    df_original, df_perplexity, on="prompt", suffixes=("_orig", "_perp")
)

df_true_filtered = merged_df[
    (merged_df["jailbroken_orig"] == True) & (merged_df["filtered"] == True)
]
df_prompt_true_filtered = df_true_filtered[["adv_prompt_perp"]]

df_true_not_filtered = merged_df[
    (merged_df["jailbroken_orig"] == True) & (merged_df["filtered"] == False)
]
df_prompt_true_not_filtered = df_true_not_filtered[["adv_prompt_perp"]]

df_false_filtered = merged_df[
    (merged_df["jailbroken_orig"] == False) & (merged_df["filtered"] == True)
]
df_prompt_false_filtered = df_false_filtered[["adv_prompt_perp"]]

df_false_not_filtered = merged_df[
    (merged_df["jailbroken_orig"] == False) & (merged_df["filtered"] == False)
]
df_prompt_false_not_filtered = df_false_not_filtered[["adv_prompt_perp"]]
```
